## Analysis Tasks :

### 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
Employee = pd.DataFrame(df["Team"].value_counts()).reset_index()
Employee
```

```python
# Added a new column relative percent to represent relative split percent of employees to total.
Total = Employee["count"].sum()
Employee["Relative percent"] = Employee["count"]/Total*100
Employee
```

```python

sns.displot(x=Employee["count"],hue=Employee["Relative percent"],bins=10,discrete=True,palette="Blues")
plt.xlabel("Number of Employees")
plt.ylabel("Number of Teams")
plt.title("Distribution of Employees across Teams",color="darkblue")
plt.xticks(np.arange(12,20))
plt.yticks(np.arange(0,25,2))

plt.show()
```

#### About 23 teams have 15 employees each , with each team having 3.27% of the company's total..Only 2 teams has 18 and 19 employees respectively which constitutes about 4% of total.In short. we can say that teams have a 3%-4% of employee distribution.

### 4. Discover which team and position have the highest salary expenditure.

```python
max_salary = df["Salary"].max()
df[df["Salary"]== max_salary][["Team","Position","Salary"]]
```

```python
#Created a dataframe for top 10 salaries with Team , Position and Salary.
TP = pd.DataFrame(df[["Team","Position","Salary"]]).sort_values("Salary",ascending=False)
Team_Position=TP.head(10)
Team_Position
```

```python
Pos=Team_Position["Position"]
sns.scatterplot(x=Team_Position["Salary"],y=Team_Position["Team"],hue=Pos)
plt.title("Top 10 Salaries By Team and Positions")
plt.show()

```

#### Employee in Team Los Angeles lakers with SF job has the highest salary. Using hues to represent position in scatterplot, it is possible to find the team and the position that has highest salaries.

### 5. Investigate if there's any correlation between age and salary, and represent it visually.

```python
Age_Salary = pd.DataFrame(df[["Age","Salary"]])
Age_Salary
```

```python
 Age_Salary.corr()
```

```python
sns.lmplot(x='Age', y='Salary', data=Age_Salary)
plt.title(' Relation Between Age And Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

#### There is a small positive correlation of 0.209124. As the age increases , there is an increase in salary which is visible from both scatterplot as well as the regression line.
